```python
fav_food_count = new_studentsurveydf["What is your favorite type of food?"].value_counts()
fav_food_count.plot(kind="pie")
plt.ylabel('Types of Food')
# This is commented out for the same reason as the chart above.

# This pie chart shows that a quarter of the class's favorite food is Italian food.
# Japanese and Mexican aren't far off, with the other ones then being smaller.
```

```python
pet_type_count = new_studentsurveydf["What kind of pets do you have?"].value_counts()
pet_type_count.plot.barh()
plt.xlabel('Amount')
plt.ylabel("Types of Pets")
# This is commented for the same reason as the two charts above.
plt.title('Student Pets')

# This graph shows that most of the students have no pets, with the second most having just dogs.
# This graph also shows none of the students just have cats, but also dogs.
```

```python
place_of_birth_count = new_studentsurveydf["What state/country were you born in?"].value_counts()
place_of_birth_count.plot(kind='pie')
plt.ylabel('Places of Birth')

# This graph shows all of the birth places from students in the class.
# The vast majority are from Missouri, with everyone else from a different state or country.
# This shows how diverse our class is, even with half being from the same state.
```
