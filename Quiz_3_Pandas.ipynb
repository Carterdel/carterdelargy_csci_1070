```python
import pandas as pd
import numpy as np

belizelistingspath = 'belize_airbnb_listings.csv'
belize_df = pd.read_csv(belizelistingspath)


grouped_belize_df = belize_df.groupby('neighbourhood')

grouped_belize_df.count()
grouped_belize_price = grouped_belize_df[['price']].agg(['mean'])
grouped_belize_reviews = grouped_belize_df[['number_of_reviews']].agg(['sum'])
grouped_belize_availability = grouped_belize_df[['availability_365']].agg(['max'])

pt1_total_belize_grouped = pd.merge(grouped_belize_price, grouped_belize_reviews, on='neighbourhood')
pt2_total_belize_grouped = pd.merge(pt1_total_belize_grouped, grouped_belize_availability, on='neighbourhood')
pt2_total_belize_grouped
```
